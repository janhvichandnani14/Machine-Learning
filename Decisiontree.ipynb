```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

 IMPORTING AND READING THE DATA

```python
df=pd.read_csv('salaries.csv')
df
```

DATA PREPROCESSING

```python
indpvar=df.drop('salary_more_then_100k',axis='columns')
target=df['salary_more_then_100k']
```

```python
indpvar.head()
```

```python
target.head()
```

```python
 le_company=LabelEncoder()
 le_job=LabelEncoder()
 le_degree=LabelEncoder()
```

```python
indpvar['company_n']=le_company.fit_transform(indpvar['company'])
indpvar['job_n']=le_company.fit_transform(indpvar['job'])
indpvar['degree_n']=le_company.fit_transform(indpvar['degree'])
indpvar.head()
```

```python
indpvar_n=indpvar.drop(['company','job','degree'],axis='columns')
indpvar_n
```

TEST-TRAIN SPLIT

```python
indpvar_n_train,indpvar_n_test,target_train,target_test=train_test_split(indpvar_n,target,test_size=0.2)
```

DECISION TREE WITH CRITERION GINI INDEX

```python
clf_gini=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=0)
clf_gini.fit(indpvar_n_train,target_train)
```

PREDICT THE TEST RESULTS

```python
Y_gini_pred=clf_gini.predict(indpvar_n_test)
```

CHECK ACCURACY SCORES

```python
print('model accuracy score with criterion gini index:{0:0.4f}'.format(accuracy_score(target_test,Y_gini_pred)))
```

```python
Y_gini_pred_train=clf_gini.predict(indpvar_n_train)
Y_gini_pred_train
```

```python
print('training set accuracy score:{0:0.4f}'.format(accuracy_score(target_train,Y_gini_pred_train)))
```

```python
print('Training set score: {:.4f}'.format(clf_gini.score(indpvar_n_train, target_train)))

print('Test set score: {:.4f}'.format(clf_gini.score(indpvar_n_test, target_test)))
```

DECISION TREE CLASSIFIER WITH CRITERION ENTROPY

```python
clf_en=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=0)
clf_en.fit(indpvar_n_train,target_train)
```

```python
y_pred_en=clf_en.predict(indpvar_n_test)
y_pred_en
```

ACCURACY SCORES

```python
print('model accuracy score with criterion entropy:{0:0.4f}'.format(accuracy_score(target_test,y_pred_en)))
```

```python
y_pred_en_train=clf_en.predict(indpvar_n_train)
y_pred_en_train
```

```python
print('training set accuracy score:{0:0.4f}'.format(accuracy_score(target_train,y_pred_en_train)))
```

```python
print('Training set score: {:.4f}'.format(clf_en.score(indpvar_n_train, target_train)))

print('Test set score: {:.4f}'.format(clf_en.score(indpvar_n_test, target_test)))
```
